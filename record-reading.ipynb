```python
import pandas as pd

# Sample data
data = pd.read_csv("MICESMOD-07.csv")
df = pd.DataFrame(data)
df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%d%H:%M:%S')
#Sorting by time
# Sort by 'time' column
dr_after_sort = df.sort_values(by='time').reset_index(drop=True)
# print(dr_after_sort)
# Initialize the new_data dictionary to store cumulative sums
new_data = {"success": 0, "fail": 0, "expire": 0, "block": 0, "process": 0}
target_sums = [17839,19133,18519,19106,19746,14995]  # List of target sums
current_target_index = 0  # To track which target sum we are working towards

# List to store results
results = []

# Loop through each row in the DataFrame
for index, row in df.iterrows():
    # Add each column value to the corresponding key in new_data
    for key in new_data.keys():
        new_data[key] += row[key]
        
    # Check if the cumulative sum has reached or exceeded the current target
    current_sum = sum(new_data.values())
   
    
    if current_sum >= target_sums[current_target_index]:
        # Print results when target is reached
        #print(f"Reached target sum {target_sums[current_target_index]} at index {index}")

        
        # Calculate the sum of fail, expire, block, and process
        total_fail_expire_block_process = new_data['fail'] + new_data['expire'] + new_data['block'] + new_data['process']
        total_dr = new_data['success'] + new_data['fail']+new_data['expire']+new_data['block']+new_data['process']
        # Store the result in the list
        results.append({
            'total dr': total_dr,
            'total fail': total_fail_expire_block_process,
            'success': new_data['success'],
            'fail': new_data['fail'],
            'expire': new_data['expire'],
            'block': new_data['block'],
            'process': new_data['process']
        })
        
        # Reset new_data for the next set of rows
        new_data = {key: 0 for key in new_data}
      
        # Move to the next target if it exists
        current_target_index += 1
        if current_target_index >= len(target_sums):
            break  # Stop if all targets have been reached

# Convert results to a DataFrame
df_results = pd.DataFrame(results)

# Save the result to a new CSV file
df_results.to_csv('target_results.csv', index=False)

#print("New file 'target_results.csv' has been created.")
```
